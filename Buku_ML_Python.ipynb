```python
#Penjumlahan angka ganjil

print("Selamat datang di program menghitung angka ganjil")

#User memasukkan angka ganjil sesuai dengan keinginan
start = int(input("masukkan angka minimum"))
end = int(input("masukkan angka maksimum"))

hasil = 0

for i in range(start,end):
    if (i%2==1):
        hasil = hasil + i

print(f"jumlah penambahan angka ganjil adalah {hasil}")
```

```python
jarak = int(input("masukkan jarak :"))
waktu = float(input("masukkan waktu (jam):"))
def kecepatan():
    hasilnya = jarak * waktu
    print(f"kecepatan kamu adalah {hasilnya} km/h")

kecepatan()
```
