```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
avg = 50
dev = 10
x1 = 45
x2 = 62

random_grade = np.random.normal(avg, dev)
mean = np.mean(random_grade)
std = np.std(random_grade)
moe1 = (x1 - avg ) / dev
moe2 = (x2 - avg ) / dev
snb1 = stats.norm.cdf(moe1)
snb2 = stats.norm.cdf(moe2)
print(moe1)
print(snb1)
print(moe2)
print(snb2)
print(snb2-snb1)
```

```python
x = np.linspace(avg - 3 * dev, avg + 3 * dev, 100)
y = stats.norm.pdf(x, avg, dev)
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Normal Distribution', color='blue')

# Highlight x1 on the plot
plt.axvline(x1, color='red', linestyle='dashed', linewidth=2, label=f'x1 = {x1}')
plt.axvline(x2, color='green', linestyle='dashed', linewidth=2, label=f'x2 = {x2}')

# Add labels and title
plt.title('Normal Distribution with Highlighted x1')
plt.xlabel('Grades')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```
